Acessado em 27/08/2025: https://www.gov.br/receitafederal/dados/arrecadacao-natureza-metadados.pdf

**Arrecadação por Natureza Jurídica e Tributo**

Contém os dados de arrecadação bruta mensal realizadas em Documento de Arrecadação de Receitas Federais (Darf), por tributo, efetuada pelos diversos tipos de Naturezas Jurídicas.

```python
# Arrecadação por Natureza Jurídica e Tributo

# Importando os dados em arquivo csv usando a biblioteca pandas
import pandas as pd

# Caminho do arquivo CSV
caminho_arquivo = r'C:\Users\merc-\\OneDrive\Documentos\Logo Ali\2025\Analise_Dados\arrecadacao-natureza.csv'  
                  # para criar uma raw string. Isso diz ao Python para ignorar as sequências de escape.
# Importar o CSV
df_natureza_juridica_e_tributo = pd.read_csv(caminho_arquivo, encoding='latin1', sep=';', low_memory=False)

# Exibir as primeiras linhas
df_natureza_juridica_e_tributo.head()
```

### Entendendo as informações dos dados de Natureza Jurídica e Tributos

```python
# Observando quantas linas e colunas
df_natureza_juridica_e_tributo.shape
```

## 1 - Análise **Descritiva** para conhecer o que está acontecendo na organização e entender tendências e causas subjacentes de tais ocorrências.

```python
# Criar um gráfico

import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(data=df_natureza_juridica_e_tributo.reset_index(), x='Contribuição Previdenciária', y='Ano')
plt.title('Contribuição Previdenciária por Tipo de Órgão Público')
plt.xlabel('Valor (R$)')
plt.ylabel('Natureza Jurídica')
plt.tight_layout()
plt.show()
```
